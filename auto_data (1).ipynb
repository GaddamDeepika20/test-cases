# Represent Corelation between variables ?

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(),annot=True,linewidths=0.5)
plt.title("correlation heatmap")
plt.show()
```

# plot different kinds of graphs for mean value represetation ?

```python
mean_values=data.mean()
mean_values
```

### Bar plot for mpgnby origin

```python
mean_mpg_by_origin=data.groupby('origin')['mpg'].mean()
```

```python
mean_mpg_by_origin.plot(kind='bar')
plt.title=("mean mpg by origin")
plt.xlabel("orogin")
plt.ylabel("mean mpg ")
plt.show()
```

### Histogram for mean weight

```python
mean_horsepower=data['weight'].mean()
plt.hist(data['weight'],bins=20,alpha=0.5,color='b',label='weight')
plt.axvline(mean_horsepower,color='r',linestyle='dashed',linewidth=2,label="mean weight")
plt.xlabel('weight')
plt.ylabel('frequency')
plt.legend()
plt.show()
```

### grouping the data by model_year and calculating the mean of the specified attributes

```python
mean_values=data.groupby('model_year')[['mpg','cylinders','displacement','horsepower','weight','acceleration']].mean()
mean_values.plot(kind='bar',subplots=True , layout=(2,3),figsize=(15,8),title="mean values by model year")
plt.tight_layout()
plt.show()
```

# identify name where acceleration is greater than 15 and horsepower less than 170 ?

```python
#convert the horsepower column to numeric values
data['horsepower']=pd.to_numeric(data['horsepower'],errors='coerce')
#filter the dataframe based on the conditions
filtered_data=data[(data['acceleration']>15)&(data['horsepower']<170)]
#display the name column of the filtered data frame
filtered_data['name']
```
